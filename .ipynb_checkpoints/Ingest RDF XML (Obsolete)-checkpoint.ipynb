```python
##Variables

sourcePath = r'C:\Users\jame9353\Box Sync\Data\NetOwl\esri_all_out2'
outPath = r'C:\Users\jame9353\Box Sync\Data\NetOwl\outCSV'
outText = os.path.join(outPath, 'NetOwl_Output.csv')
```

```python
#Create a Graph Object
g = Graph()
```

```python
#Walk through Source Path and Parse all RDF/XML Documents
for root, dir, files in os.walk(sourcePath):
    for file in files:
        if file.endswith('.rdf'):
            filePath = os.path.join(root, file)
            print("Parsing " + file + "...")
            g.parse(filePath, format='xml')
```

```python
#Iterate through Graph object and parse Triples.  Writes Triples to a text file.
print("Parsing Graph into triples and writing to CSV")
with open(outText, 'w', newline='\n') as csvFile:
    csvWriter = csv.writer(csvFile, delimiter=',')
    totalTriples = 0
    error = 0
    subjects = []
    predicates = []
    objs = []
    for subject, predicate, obj in g:
        totalTriples +=1
        if subject not in subjects:
            subjects.append(subject)

        if predicate not in predicates:
            predicates.append(predicate)

        if obj not in objs:
            objs.append(obj)

        cleanSubj = os.path.split(subject)[1]
        cleanPred = predicate.split('#')[1]
        cleanObj = obj[0]
        
        try:
            csvWriter.writerow(([cleanSubj, cleanPred, obj]))
        except:
            print("Error on writing triple")
            print(cleanTrip)
            error += 1
            if float(error) / float(totalTriples) * 100 >= 2:
                break
                print("Too many errors with documents...")
```
