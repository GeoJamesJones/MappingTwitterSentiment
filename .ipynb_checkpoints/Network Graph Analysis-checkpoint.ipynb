```python
import csv
import os
import networkx as nx
import pandas as pd
from operator import itemgetter
```

```python
docsPath = r'C:\Users\jame9353\Documents\GitHub\RDFtoNeo4J\Demo\Gang Network'
file = 'SourceGangNetwork.csv'

csvFile = os.path.join(docsPath, file)
```

```python
df = pd.read_csv(csvFile)
```

```python
df
```

```python
G = nx.Graph()
nx.from_pandas_dataframe(df, source='Name', target='Called Name', edge_attr='Contact Type', create_using=G)
```

```python
name = {}
affiliation = {}
role = {}
```

```python
for row in df.iterrows():
    if row[1]['Name'] not in name.keys():
         name[row[1]['Name']] = row[1]['Phone Number']
    if row[1]['Name'] not in affiliation.keys():
         affiliation[row[1]['Name']] = row[1]['Affiliation']
    if row[1]['Name'] not in role.keys():
         role[row[1]['Name']] = row[1]['Role']
```

```python
nx.set_node_attributes(G, name, 'Name')
nx.set_node_attributes(G, affiliation, 'Affiliation')
nx.set_node_attributes(G, role, 'Role')
```

```python
for n in G.nodes():
    try: print(n, G.node[n]['Name'])
    except:pass
```

```python
pos = nx.spring_layout(G,scale=1)
```

```python
print(nx.info(G))
```

```python
density = nx.density(G)
print("Network density:", density)
```

```python
print(G.nodes())
```

```python
sourceNode = 'Scott Gray'
targetNode = 'Juan Carlos'
shortestPath = nx.shortest_path(G, source=sourceNode, target=targetNode)
print("Shortest Path between nodes:  ", shortestPath)
```

```python
components = nx.connected_components(G)
largest_component = max(components, key=len)
print(largest_component)
```

```python
nx.draw(G, with_labels = True)
```

```python
from operator import itemgetter
degree_dict = dict(G.degree(G.nodes()))
nx.set_node_attributes(G, degree_dict, 'degree')
sorted_degree = sorted(degree_dict.items(), key=itemgetter(1), reverse=True)
print("Top 5 nodes by degree:")
for d in sorted_degree[:5]:
    print(d)
```

```python
betweenness_dict = nx.betweenness_centrality(G) # Run betweenness centrality
eigenvector_dict = nx.eigenvector_centrality(G) # Run eigenvector centrality
nx.set_node_attributes(G, betweenness_dict, 'betweenness')
nx.set_node_attributes(G, eigenvector_dict, 'eigenvector')

sorted_betweenness = sorted(betweenness_dict.items(), key=itemgetter(1), reverse=True)

#First get the top 5 nodes by betweenness as a list
top_betweenness = sorted_betweenness[:5]

#Then find and print their degree
for tb in top_betweenness: # Loop through top_betweenness
    degree = degree_dict[tb[0]] # Use degree_dict to access a node's degree
    print("Name:", tb[0], "| Betweenness Centrality:", tb[1], "| Degree:", degree)
```

```python
triadic_closure = nx.transitivity(G)
print("Triadic closure:", triadic_closure)
```

```python
A = list(nx.find_cliques(G))
for i in A:
    print(i)
```

```python
bridges = list(nx.bridges(G))
for b in bridges:
    print(b)
```
